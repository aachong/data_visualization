```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
data = pd.read_csv('data/3d-line1.csv')
data.head()
```

## 通过go来画3D图
* mode='lines'只画线
* mode='markers'只画点
* marger = {'size':3} 调整点的大小

```python
line1 = go.Scatter3d(x=data['x'],y=data['y'],z=data['z'],mode='lines')
fig1 = go.Figure(line1)
fig1.show()
```

```python
fig2 = px.scatter_3d(data,x='x',y='y',z='z',color='color')
fig2.show()
```
