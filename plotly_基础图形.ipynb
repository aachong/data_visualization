```python
import pandas as pd
import plotly
import plotly.graph_objects as go

data = pd.read_csv('data/2016-weather-data-seattle.csv')
data.head(),data['Date'].shape,data['Max_TemperatureC'].shape
data['Date'][:3]
```

### Scatter散点图

```python
line1 = go.Scatter(x=data['Date'][:100],y=data['Max_TemperatureC'][:100],name='max')
line2 = go.Scatter(x=data['Date'][:100],y=data['Mean_TemperatureC'][:100],name='mean')
fig = go.Figure([line1,line2])
fig.update_layout(title='good',xaxis_title='x',yaxis_title='y')
fig.show()
```

### Bar条形图

```python
line3 = go.Bar(x=data['Date'][:10],y=data['Max_TemperatureC'][:10],name='max')
line4 = go.Bar(x=data['Date'][:10],y=data['Mean_TemperatureC'][:10],name='mean')
fig2 = go.Figure([line3,line4])
fig2.show()
```

### Histogram直方图（y坐标为x坐标的个数）

```python
line5 = go.Histogram(x=data['Max_TemperatureC'])
fig3 = go.Figure(line5)
fig3.update_layout(bargap=0.1) #添加间隙
fig3.show()
```

```python
line5 = go.Histogram(x=data['Max_TemperatureC'],xbins={'size':3})#添加统计间隔，每隔都少个加一个
fig3 = go.Figure(line5)
fig3.update_layout(bargap=0.1) #添加间隙
fig3.show()
```
